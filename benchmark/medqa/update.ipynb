```python
import numpy as np

import json
import re
from sklearn.metrics import roc_auc_score
```

```python
def extract_confidence(text: str) -> float | None:
    """
    从文本中提取“置信度”后面的第一个浮点数，返回一个 float，
    如果没找到则返回 None。
    """
    # 匹配“置信度：”或“置信度:”，后面跟着可选空白，再跟一个整数或浮点数
    pattern = r"置信度[\s\S]*?([0-9]+(?:\.[0-9]+)?)"
    match = re.search(pattern, text)
    if match:
        return float(match.group(1))
    return None
```

```python
def update(model_name: str):
    # 读取json文件
    with open(f"{model_name}/results_calibrate.json", "r") as f:
        data = json.load(f)
    raw_confidence = data["raw_confidences"]

    accs = []
    raw_confidences = []
    for a, b in zip(data["acc"], raw_confidence):
        b = extract_confidence(b)
        if b is not None:
            accs.append(a)
            raw_confidences.append(b)
    print(accs[-2:])
    print(raw_confidences[-2:])
    updated_auroc = roc_auc_score(np.array(accs), np.array(raw_confidences))
    print(f"{model_name} updated auroc: {updated_auroc}")
```

```python
update("Qwen2.5-7B-Instruct")
```
