```python
import matplotlib.pyplot as plt
import os
import json
from typing import Optional, List
import random

results_dir = "results"
```

```python
def plot_logprobs(model_name, dataset_name, selection: Optional[List[int]] = None):
    with open(
        os.path.join(results_dir, model_name, dataset_name, "calibration.json")
    ) as f:
        data = json.load(f)

    if selection is not None:
        figsize = (5, 3)
        num_plots = len(selection)
        fig, axes = plt.subplots(
            nrows=num_plots,
            ncols=1,
            squeeze=False,
            sharey=True,
            sharex=True,
            layout="constrained",
            figsize=(figsize[0] * num_plots, figsize[1] * num_plots),
        )
        for i, idx in enumerate(selection):
            ax = axes[i, 0]
            ax.plot(data[idx])
            ax.set_xlabel(f"Sample {idx}")
            ax.set_ylabel("Logprobs")
```

```python
indices = random.sample(list(range(1000)), 5)
plot_logprobs("Qwen2.5-7B-Instruct", dataset_name="gsm8k", selection=indices)
```
